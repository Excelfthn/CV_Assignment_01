```python
import matplotlib.pyplot as plt
```

```python
# Dummy blurred grayscale image (5x5 matrix)
blurred_image = [
    [60, 62, 63, 61, 60],
    [61, 64, 65, 64, 62],
    [62, 65, 66, 65, 63],
    [61, 64, 65, 64, 62],
    [60, 62, 63, 62, 60]
]

print("Blurred Image Matrix:")
for row in blurred_image:
    print(row)

# Visualize
plt.imshow(blurred_image, cmap='gray', vmin=0, vmax=255)
plt.title("Blurred Image (Dummy)")
plt.colorbar()
plt.show()
```

```python
# Sharpening kernel
kernel = [
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
]

rows = len(blurred_image)
cols = len(blurred_image[0])

# Create output matrix
sharpened = [[0 for _ in range(cols)] for _ in range(rows)]

# Manual convolution
for i in range(1, rows-1):
    for j in range(1, cols-1):
        value = 0
        for ki in range(-1, 2):
            for kj in range(-1, 2):
                value += blurred_image[i+ki][j+kj] * kernel[ki+1][kj+1]

        # clamp value
        if value < 0:
            value = 0
        if value > 255:
            value = 255

        sharpened[i][j] = value

print("Sharpened Image Matrix:")
for row in sharpened:
    print(row)

# Visualization
plt.imshow(sharpened, cmap='gray', vmin=0, vmax=255)
plt.title("Sharpened Image")
plt.colorbar()
plt.show()
```

```python
dark_image = [
    [5, 10, 12, 8, 6],
    [7, 12, 15, 11, 9],
    [8, 14, 18, 13, 10],
    [7, 12, 15, 11, 9],
    [6, 9, 12, 10, 7]
]

print("Dark Image Matrix:")
for row in dark_image:
    print(row)

plt.imshow(dark_image, cmap='gray', vmin=0, vmax=255)
plt.title("Dark Image (Dummy)")
plt.colorbar()
plt.show()
```

```python
import math

max_pixel = max(max(row) for row in dark_image)
c = 255 / math.log(1 + max_pixel)

log_dark = []

for row in dark_image:
    new_row = []
    for pixel in row:
        value = c * math.log(1 + pixel)
        value = int(value)

        if value > 255:
            value = 255

        new_row.append(value)
    log_dark.append(new_row)

print("Enhanced Dark Image Matrix:")
for row in log_dark:
    print(row)

plt.imshow(log_dark, cmap='gray', vmin=0, vmax=255)
plt.title("Dark Image Enhancement (Log Transform)")
plt.colorbar()
plt.show()
```
